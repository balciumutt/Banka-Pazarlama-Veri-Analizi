### Fiyat Stratejisi Karar Destek

```python
X = df[['balance', 'duration']]
X = sm.add_constant(X)
y = df['deposit'].apply(lambda x: 1 if x == 'yes' else 0)
model = sm.Logit(y, X)
results = model.fit()
#
print(results.summary())
```

### Olasılık Dağılımları

#### Bernoulli Dağılımı

```python
p_default = df['default'].value_counts(normalize=True)['yes']
bernoulli_dist = bernoulli(p_default)

print("P(X=0) =", bernoulli_dist.pmf(0))
print("P(X=1) =", bernoulli_dist.pmf(1))
```

#### Binom Dağılımı

```python
n_campaign = len(df)  
p_campaign = df['campaign'].mean()  
binom_dist = binom(n_campaign, p_campaign)
k = 5
print("P(X <= {}) =".format(k), binom_dist.cdf(k))
```

#### Geometrik Dağılım

```python
p_pdays = df['pdays'].value_counts(normalize=True).iloc[0] 
geometric_dist = geom(p_pdays)
k = 1
print("P(X = {}) =".format(k), geometric_dist.pmf(k))
```

#### Poisson Dağılımı

```python
lambda_previous = df['previous'].mean() 
poisson_dist = poisson(mu=lambda_previous)
k = 2
print("P(X = {}) =".format(k), poisson_dist.pmf(k))
```

#### Normal Dağılım

```python
mean_balance = df['balance'].mean()
std_balance = df['balance'].std()
normal_dist = norm(loc=mean_balance, scale=std_balance)
value = 1500
print("P(X <= {}) =".format(value), normal_dist.cdf(value))
```

#### Üssel Dağılım

```python
def expon_pdf(x, lambd):
    return lambd * np.exp(-lambd * x)
lambd = 0.5
value = 1
print("P(X <= {}) =".format(value), expon_pdf(value, lambd))
```

### Hipotez Testi

#### Tek Örneklem T Testi

```python
stat, p_value = ttest_1samp(df['balance'], popmean=1500)
print("Tek Örneklem T Testi")
print("T İstatistiği =", stat)
print("P değeri =", p_value)
```

#### Bağımsız İki Örneklem T Testi

```python
from scipy.stats import ttest_ind
group_A = df[df['marital'] == 'married']['balance']
group_B = df[df['marital'] == 'single']['balance']

stat, p_value = ttest_ind(group_A, group_B)

print("Bağımsız İki Örneklem T Testi (AB Testi)")
print("T İstatistiği =", stat)
print("P değeri =", p_value)
```

#### Bağımlı İki Örneklem T Testi

```python
from scipy.stats import ttest_rel
stat, p_value = ttest_rel(df['balance'], df['balance'] + df['previous'])
print("Bağımlı İki Örneklem T Testi")
print("T İstatistiği =", stat)
print("P değeri =", p_value)
```

#### İki Örneklem Oran Testi

```python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

df['deposit_numeric'] = df['deposit'].apply(lambda x: 1 if x == 'yes' else 0).astype(int)

group_A = df[df['marital'] == 'married']['deposit_numeric']
group_B = df[df['marital'] == 'single']['deposit_numeric']

success_A = group_A.sum()
success_B = group_B.sum()

n_A = len(group_A)
n_B = len(group_B)

count = np.array([success_A, success_B])
nobs = np.array([n_A, n_B])

stat, p_value = proportions_ztest(count, nobs)

print("İki Örneklem Oran Testi")
print("Z İstatistiği =", stat)
print("P değeri =", p_value)
```

#### Varyans Analizi

```python
from scipy.stats import f_oneway

groups = [df[df['job'] == job]['balance'] for job in df['job'].unique()]
stat, p_value = f_oneway(*groups)

print("Varyans Analizi (ANOVA)")
print("F İstatistiği =", stat)
print("P değeri =", p_value)
```

#### Korelasyon Analizi

```python
from scipy.stats import pearsonr

correlation, p_value = pearsonr(df['balance'], df['duration'])

print("Korelasyon Analizi")
print("Korelasyon Katsayısı =", correlation)
print("P değeri =", p_value)
```

### Tüm Modellerin Karşılaştırması

```python
modeller = [
    knn_tuned,
    loj_model,
    nb_model,
    mlpc_tuned,
    cart_tuned,
    rf_tuned,
    gbm_tuned,
    catb_tuned,
    lgbm_tuned,
    xgb_tuned
    
]


for model in modeller:
    if(model!=mlpc_tuned):
        y_pred = model.predict(X_test)
    else:
        y_pred = model.predict(X_test_scaled)
        
    dogruluk = accuracy_score(y_test, y_pred)
    isimler = model.__class__.__name__
    print("-"*30)
    print(isimler + ":" )
    print("Accuracy: {:.4%}".format(dogruluk))
```

```python
sonuclar = pd.DataFrame(columns=["modeller", "Accuracy"])

for model in modeller:
    if model != mlpc_tuned:
        y_pred = model.predict(X_test)
    else:
        y_pred = model.predict(X_test_scaled)

    dogruluk = accuracy_score(y_test, y_pred)
    isimler = model.__class__.__name__
    sonuc = pd.DataFrame([[isimler, dogruluk * 100]], columns=["modeller", "Accuracy"])
    sonuclar = pd.concat([sonuclar, sonuc], ignore_index=True)

sns.barplot(x='Accuracy', y='modeller', data=sonuclar, color="r")
plt.xlabel('Accuracy %')
plt.title('Modellerin Doğruluk Oranları')
plt.show()
```

```python
# Save Model Using Pickle
import pandas
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
import pickle
#uzak sunucudan datanin cekilmesi
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
dataframe = pandas.read_csv(url, names=names)
array = dataframe.values
X = array[:,0:8]
Y = array[:,8]
test_size = 0.33
seed = 7
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=test_size, random_state=seed)
# Fit the model on training set
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
# save the model to disk
filename = 'data/finalized_model.sav'
pickle.dump(model, open(filename, 'wb'))
```

```python
# load the model from disk
loaded_model = pickle.load(open(filename, 'rb'))
result = loaded_model.score(X_test, Y_test)
print(result)
```
